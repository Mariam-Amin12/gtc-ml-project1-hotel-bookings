# identify and handle outliers

```python
def removal_box_plot(df, column, threshold):
    removed_outliers = df[df[column] <= threshold]

    sns.boxplot(removed_outliers[column])
    plt.title(f'Box Plot without Outliers of {column}')
    plt.show()
    return removed_outliers



```

```python

# Get numerical columns
numerical_features = df.select_dtypes(include=[np.number]).columns

# Apply boxplot and outlier removal to all numerical features
for feature in numerical_features:
    df_diabetics = pd.DataFrame(df, columns=df.columns)
    
    sns.boxplot(df_diabetics[feature])
    plt.title(f'Boxplot of {feature}')
    plt.show()
    
   

```

# Feature Engineering & Preprocessing



```python
print("Creating new features...")

df['total_guests'] = df['adults'] + df['children'] 



df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

df['is_family'] = (df['children'] > 0).astype(int)
```

```python
# One-Hot Encoding for low-cardinality categorical columns
low_cardinality_cols = ['hotel', 'meal', 'market_segment', 'distribution_channel', 
                        'reserved_room_type', 'deposit_type', 'customer_type']

df_encoded = pd.get_dummies(df, columns=low_cardinality_cols, drop_first=False)

# Frequency encoding for high-cardinality columns
for col in ['country', 'assigned_room_type', 'arrival_date_month']:
    freq_encoding = df[col].value_counts().to_dict()
    df_encoded[f'{col}_frequency'] = df[col].map(freq_encoding)

# Grouping infrequent categories into 'Other' for high-cardinality columns
def group_infrequent(series, min_count=200):
    freq = series.value_counts()
    to_keep = freq[freq >= min_count].index
    return series.where(series.isin(to_keep), other='Other')

df_encoded['country_grouped'] = group_infrequent(df['country'])
df_encoded['assigned_room_type_grouped'] = group_infrequent(df['assigned_room_type'])
df_encoded['arrival_date_month_grouped'] = group_infrequent(df['arrival_date_month'])

# Optionally, one-hot encode the grouped columns if needed
df_encoded = pd.get_dummies(df_encoded, columns=['country_grouped', 'assigned_room_type_grouped', 'arrival_date_month_grouped'], drop_first=False)
```

```python
#drop the not needed columns
df_encoded.drop(columns=['country', 'assigned_room_type', 'arrival_date_month'], inplace=True)
```

```python
# Show correlation of each feature with target variable (is_canceled)
print("Feature Correlations with 'is_canceled':")
print("=" * 40)

# Get numerical features only
numerical_features = df.select_dtypes(include=[np.number]).columns

# Calculate correlation with target variable
for feature in numerical_features:
    if feature != 'is_canceled':
        corr = df[feature].corr(df['is_canceled'])
        print(f"{feature:<30}: {corr:.3f}")

# Show correlation matrix heatmap
plt.figure(figsize=(12, 10))
corr_matrix = df[numerical_features].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()
```
